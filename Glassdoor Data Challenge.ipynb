## A. Categorical Predictor

```python
def plot_win_pct(bycat, data=datatrain, hori=True):
    """
    plot winning percentage by category variable bycat
    
    """
    df0 = datatrain[['isWon', bycat]].groupby(bycat).agg('mean')
    df0.columns = ['win_pct']
    df0 = df0.sort_values('win_pct', ascending=False)
    df0[bycat] = df0.index
    df0.index = range(df0.shape[0])
    
    if hori == True:
        sns.barplot(y=bycat, x="win_pct", data=df0, color="cornflowerblue")
        plt.axvline(x=datatrain['isWon'].mean(), color='red')
    else:
        sns.barplot(x=bycat, y="win_pct", data=df0, color="cornflowerblue")
        plt.axhline(y=datatrain['isWon'].mean(), color='red')
        
def plot_win_pct_byconti(byconti, data=datatrain):
    df0 = data[['isWon', byconti]].groupby(byconti).agg('mean')
    df0.columns = ['win_pct']
    df0 = df0.sort_values('win_pct', ascending=False)
    df0[byconti] = df0.index
    df0.index = range(df0.shape[0])
    
    sns.lmplot(x=byconti, y='win_pct', data=df0)
```

# Model Building

## 1. Label Encoding

Convert categorical predictor to integers.

```python
from sklearn.preprocessing import LabelEncoder
```

```python
datatrain.columns.to_series().groupby(datatrain.dtypes).groups
```

```python
cat_col = ['ATS__c', 'Industry']
```

```python
LabelMap = dict()
for i in cat_col:
    print('Start encoding ' + i)
    df_tmp = pd.concat([datatrain[i], datarun[i]])
    df_tmp = df_tmp.fillna('missing')
    le = LabelEncoder()
    le.fit(df_tmp)
    LabelMap[i] = dict(zip(le.classes_, le.transform(le.classes_)))
```

```python
for i in LabelMap:
    LabelMap[i]['missing'] = np.nan

LabelMap
```

```python
# Save Label Mapping
np.save('./input/LabelMapping.npy', LabelMap) 

# Load Label Mapping
# LabelMap = np.load('LabelMapping.npy').item()
```

```python
datatrain.replace(LabelMap, inplace=True)
datarun.replace(LabelMap, inplace=True)
```

## 3. Train/Test Data Split

```python
from sklearn.model_selection import train_test_split
```

For the Train/Test Data Split, I use a 75%/25% ratio and stratified on "Industry".

```python
train, test = train_test_split(datatrain, test_size=.25, train_size=.75, stratify=datatrain['Industry'].tolist(), random_state=111)
print(train.shape, test.shape)
```

## A. LightGBM Model

I use [LightGBM](https://github.com/Microsoft/LightGBM) as the first model. LightGBM is a fast, distributed, high performance gradient boosting framework based on decision tree algorithms, used for ranking, classification and many other machine learning tasks. I use 5-fold cross validation to tune the parameters at first.

```python
import lightgbm as lgb
from sklearn.model_selection import GridSearchCV
```

```python
# create dataset for lightgbm
X_train = train.drop(['isWon','employerId'], axis=1)
y_train = train['isWon']
X_test = test.drop(['isWon','employerId'], axis=1)
y_test = test['isWon']
lgb_train = lgb.Dataset(X_train, y_train, categorical_feature=cat_col)
lgb_eval = lgb.Dataset(X_test, y_test, categorical_feature=cat_col, reference=lgb_train)
```

```python
params = {
    'boosting_type': 'gbdt',
    'objective': 'binary',
    'metric': {'binary_logloss', 'auc'},
    'num_boost_round': 500,
    'learning_rate': .1,
    'num_leaves': 31,
    'max_bin': 255,
    'lambda_l2': .05,
    'num_threads':8
}

param_grid = {
    'boosting_type': ['gbdt'],
    'objective': ['binary'],
    'metric': ['binary_logloss'],
    'num_boost_round': [500],
    'learning_rate': [.1],
    'num_leaves': [20,30,50,70,90],
    'max_bin': [255,383,511],
    'lambda_l2': [0, .05, .1],
    'num_threads': [8]
}

mod = lgb.LGBMClassifier(boosting_type=params['boosting_type'],
                         objective=params['objective'],
                         metric=params['metric'],
                         num_boost_round=params['num_boost_round'],
                         learning_rate=params['learning_rate'],
                         num_leaves=params['num_leaves'],
                         max_bin=params['max_bin'],
                         lambda_l2=params['lambda_l2'],
                         num_threads=params['num_threads']
                       )
```

```python
%time
print('CV training start...')
grid = GridSearchCV(mod, param_grid, cv=5)
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
%%time
# specify your configurations as a dict
params = {
    'task': 'train',
    'boosting_type': 'gbdt',
    'objective': 'binary',
    'metric': {'binary_logloss', 'auc'},
    'num_leaves': 30,
    'max_bin': 511,
    'learning_rate': 0.01,
    'verbose': 0,
    'num_threads':8,
    'early_stopping_round':100
}

print('Start training...')
# train
gbm = lgb.train(params,
                lgb_train,
                categorical_feature = cat_col,
                num_boost_round=10000,
                valid_sets=lgb_eval)

print('Save model...')
# save model to file
gbm.save_model('lgb_model.txt')
```

```python
# predict
y_pred_lgb = gbm.predict(X_test, num_iteration=gbm.best_iteration)

print('Test accuracy is: ', np.mean(y_test == (y_pred_lgb > .5)))
```

### Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import itertools
```

```python
def plot_confusion_matrix(y_test, y_pred, classes,
                          normalize=True,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    cm = confusion_matrix(y_test, y_pred)
    
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    #plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

```python
plt.figure(figsize=(9, 9))
plot_confusion_matrix(y_test, y_pred_lgb > .3, ['notWon', 'isWon'], title="Confusion Matrix for LightGBM")
```

## B. Logistic Regression on Non-missing Features

Logistic Regression cannot work with missing values. We fit a logistic regression with non-missing features.

```python
from sklearn.linear_model import LogisticRegression
```

```python
# convert back Label Encoding
inv_LabelMap = dict()
for i in LabelMap.keys():
    inv_LabelMap[i] = {v: k for k, v in LabelMap[i].items()}
```

```python
X_train.replace(inv_LabelMap, inplace=True)
X_test.replace(inv_LabelMap, inplace=True)
```

```python
Xtrain_logi = X_train.drop(['ATS__c','Industry'], axis=1).join(pd.get_dummies(X_train['ATS__c'], prefix='ATS__c')).join(pd.get_dummies(X_train['Industry'], prefix='Industry'))
Xtest_logi = X_test.drop(['ATS__c','Industry'], axis=1).join(pd.get_dummies(X_test['ATS__c'], prefix='ATS__c')).join(pd.get_dummies(X_test['Industry'], prefix='Industry'))
```

```python
missing_col = datatrain.apply(lambda x: sum(x.isna()), axis=0).sort_values()[datatrain.apply(lambda x: sum(x.isna()), axis=0).sort_values() > 0].index.tolist()
```

```python
logi = LogisticRegression(max_iter=10000)
logi.fit(Xtrain_logi.drop(missing_col, axis=1), y_train)
```

```python
y_pred_logi1 = logi.predict_proba(Xtest_logi.drop(missing_col, axis=1))[:,1]
print('Test Accuracy is: ', np.mean(y_test == (y_pred_logi1 > .5)))
```

### The result is terrible! Therefore, we decide to impute the missing values and use all the features. We use median value as the strategy.

```python
from sklearn.preprocessing import Imputer
my_imputer = Imputer(strategy='mean')
```

```python
new_Xtrain = Xtrain_logi.copy()
new_Xtest = Xtest_logi.copy()

for col in missing_col:
    new_Xtrain[col + '_was_missing'] = new_Xtrain[col].isnull()
    new_Xtest[col + '_was_missing'] = new_Xtest[col].isnull()

new_Xtrain = my_imputer.fit_transform(new_Xtrain)
new_Xtest = my_imputer.fit_transform(new_Xtest)
```

```python
logi = LogisticRegression(max_iter=10000)
logi.fit(new_Xtrain, y_train)
```

### Logistic Regression with imputing missing values still give us a terrible performance. The accuracy is basically predicting all cases as True cases.

```python
y_pred_logi2 = logi.predict_proba(new_Xtest)[:,1]
print('Test Accuracy is: ', np.mean(y_test == (y_pred_logi2 > .5)))
```
